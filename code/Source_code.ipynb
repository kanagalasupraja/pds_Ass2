```python
df = pd.DataFrame(raw_train)
```

```python
#(a)addressed the replacement of missing values in the 'Mileage', 'Engine', 'Power', 'Seats', and 'New_Price' columns. To handle these missing values:
#For 'Mileage', 'Engine', and 'Power', we opt for the median to mitigate the impact of outliers.
#Regarding 'Seats', which only has two distinct values (5 or 7), we utilize the mode for replacement.
#dropped the 'New_Price' column altogether due to the excessive number of missing values it contains.
```

```python
#b)mileage, engine, power and new price,has units, first we will remove these
df.head()
```

```python
df['Mileage'] = df['Mileage'].str.extract('(\d+\.\d+)').astype(float)

df['Engine'] = df['Engine'].str.replace(' CC', '').astype(float)

df['Power'] = df['Power'].str.extract('(\d+\.\d+)').astype(float)

df['New_Price'] = df['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
#Replace the NA values of Mileage, Engine and Power with the mean
df['Mileage'].fillna(df['Mileage'].median(),inplace=True)
df['Engine'].fillna(df['Engine'].median(),inplace=True)
df['Power'].fillna(df['Power'].median(),inplace=True)

df['Seats'].fillna(df['Seats'].mode()[0],inplace=True)

df.drop('Price', axis=1, inplace=True)
```

```python
df.head()
```

```python
output_csv_file = '/content/clean_data.csv'
df.to_csv(output_csv_file, index=False)
```
